```python
import os

folder_path = "/content/rag_data"

documents = []

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as f:
            content = f.read()
            documents.append({"filename": filename, "text": content})

len(documents), documents[:2]  # preview 2 docs
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Initialize the Sentence Transformer model
# You can choose a different model if needed, e.g., 'all-MiniLM-L6-v2'
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

def embed_text(text):
    embeddings = embedding_model.encode([text])
    return np.array(embeddings[0])

# test embedding
test_vector = embed_text("This is a test.")
print(f"Embedding length: {len(test_vector)}")
```

```python
import faiss
import numpy as np

# build list of document embeddings
doc_embeddings = []
doc_texts = []

for doc in documents:
    vec = embed_text(doc["text"])
    doc_embeddings.append(vec)
    doc_texts.append(doc["text"])

embedding_matrix = np.vstack(doc_embeddings).astype("float32")

index = faiss.IndexFlatL2(embedding_matrix.shape[1])
index.add(embedding_matrix)

print("FAISS index size:", index.ntotal)
```

```python
def retrieve(query, k=3):
    q_vec = embed_text(query).astype("float32").reshape(1, -1)
    distances, indices = index.search(q_vec, k)

    results = []
    for i in indices[0]:
        results.append(doc_texts[i])
    return results

def groq_answer(query, context):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "system", "content": "You are a helpful RAG assistant. Use the provided context ONLY."},
            {"role": "user", "content": f"Context:\n\n{context}\n\nQuestion: {query}"}
        ]
    )
    return response.choices[0].message.content

def rag(query):
    retrieved_docs = retrieve(query, k=3)
    context = "\n\n".join(retrieved_docs)
    return groq_answer(query, context)
```
